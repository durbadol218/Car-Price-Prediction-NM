# Data Preprocessing

## 1. Null Check and Handling

```python
def clean_data(val):
    try:
        val = str(val).split(' ')[0].strip()
        return float(val) if val else np.nan
    except:
        return np.nan

cars_data['mileage'] = cars_data['mileage'].apply(clean_data)
cars_data['engine'] = cars_data['engine'].apply(clean_data)
cars_data['max_power'] = cars_data['max_power'].apply(clean_data)

# Fill missing values
cars_data['mileage'].fillna(cars_data['mileage'].mean(), inplace=True)
cars_data['engine'].fillna(cars_data['engine'].mean(), inplace=True)
cars_data['max_power'].fillna(cars_data['max_power'].mean(), inplace=True)
cars_data['seats'].fillna(cars_data['seats'].mode()[0], inplace=True)

cars_data.shape
```

## 2. Duplicate Check

```python
cars_data.duplicated().sum()
```

```python
cars_data.drop_duplicates(inplace=True)
```

```python
cars_data['name'] = cars_data['name'].replace(
    ['Maruti', 'Skoda', 'Honda', 'Hyundai', 'Toyota', 'Ford', 'Renault', 'Mahindra', 'Tata',
     'Chevrolet', 'Datsun', 'Jeep', 'Mercedes-Benz', 'Mitsubishi', 'Audi', 'Volkswagen', 'BMW',
     'Nissan', 'Lexus', 'Jaguar', 'Land', 'MG', 'Volvo', 'Daewoo', 'Kia', 'Fiat', 'Force',
     'Ambassador', 'Ashok', 'Isuzu', 'Opel'],
    list(range(1, 32))
)
```

```python
cars_data['transmission'] = cars_data['transmission'].replace(['Manual', 'Automatic'], [1, 2])
```

```python
cars_data['owner'] = cars_data['owner'].replace(['First Owner', 'Second Owner', 'Third Owner',
       'Fourth & Above Owner', 'Test Drive Car'],[1, 2, 3, 4, 5])
```

```python
cars_data
```

```python
cars_data.drop(columns=['index'], inplace=True)
```

```python
cars_data['name'] = pd.to_numeric(cars_data['name'], errors='coerce')
cars_data['name'].fillna(0, inplace=True)
cars_data['name'] = cars_data['name'].astype(int)

input_data = cars_data.drop(columns=['selling_price']) #Other 11 columns are input feature, except this "selling_price" column!
output_data = cars_data['selling_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2)
```

# Model Creation

```python
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(),
    'Decision Tree': DecisionTreeRegressor(),
}
if not os.path.exists("models"):
    os.makedirs("models")
```

# Train Model

```python
for name,model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"{name} RÂ² score: {r2_score(y_test, y_pred):.4f}")
    print(f"{name} RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.2f}")
    with open(f"models/{name.replace(' ','_').lower()}_model.pkl", "wb") as f:
        pkl.dump(model, f)
```
